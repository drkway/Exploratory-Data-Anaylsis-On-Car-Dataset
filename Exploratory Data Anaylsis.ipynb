```python
#Converting Price From Object To Int
data['price'] = pd.to_numeric(data['price'], errors='coerce')

# Calculate the median of the numeric values in the 'Price' column
price_median = data['price'].median()

# Replace non-numeric values with the median
data['price'].fillna(price_median, inplace=True)

# Convert the 'Price' column to integer type
data['price'] = data['price'].astype('int64')




#Converting HP From Object To Int
# Remove non-numeric characters from the 'HP' column and convert to numeric
data['HP'] = pd.to_numeric(data['HP'], errors='coerce')

# Calculate the median of the numeric values in the 'HP' column
hp_median = data['HP'].median()

# Replace non-numeric values with the median
data['HP'].fillna(hp_median, inplace=True)

# Convert the 'HP' column to integer type
data['HP'] = data['HP'].astype('int64')
```

```python
#Removing Outliers

# Define columns to convert to numeric
columns_to_convert = ['normalized-losses', 'bore', 'stroke', 'peak-rpm', 'price',]

# Convert each column to numeric
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Drop columns with non-numeric values
data = data.dropna(axis=1, how='any')

data.info()
data.head()
```

```python
data.info()
```

```python
# Calculate quartiles for the selected columns
columns_of_interest = ['wheel-base','length','width','height','engine-size','HP','MPG-C','MPG-H','price']
Q1 = data[columns_of_interest].quantile(0.25)
Q3 = data[columns_of_interest].quantile(0.75)

# Calculate IQR for the selected columns
IQR = Q3 - Q1
print(IQR)
```

```python
# Select only numerical columns
numerical_columns = data.select_dtypes(include=['number'])

# Ensure alignment of indices and columns
numerical_columns, Q1 = numerical_columns.align(Q1, axis=1, join='inner')
numerical_columns, Q3 = numerical_columns.align(Q3, axis=1, join='inner')

# Filter data to keep only rows within the interquartile range for numerical columns
data = data[~(((numerical_columns < (Q1 - 1.5 * IQR)) | (numerical_columns > (Q3 + 1.5 * IQR))).any(axis=1))]
```

```python
plt.rcParams["figure.figsize"] = (8,4)  
abc=data.copy()
```

```python
columns_to_drop = ['symboling', 'make','engine-location', 'price_grp', 
                   'aspiration', 'body-style', 'Drive Mode', 
                    'fuel-system','price','curb-weight','width','compression-ratio','engine-type','engine-size']

# Drop the specified columns
abc.drop(columns=columns_to_drop, inplace=True)

# Check the resulting DataFrame
print(abc.head())  # Display the first few rows of the DataFrame after dropping columns
```

```python
abc=pd.get_dummies(abc,drop_first=True) #to conver categorical varibles .
x=abc.iloc[:,:-1].values
y = data['price'].values.reshape(-1,1)
```

```python
# Feature Scaling (Feature scaling is a method used to normalize the range of independent variables or features of data.)
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(x)
y = sc_y.fit_transform(y.reshape(-1,1))
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#using linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
plt.scatter(y_test,y_pred)
```

```python
sns.distplot((y_test-y_pred),bins=50)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn import metrics
import math
print("Mean Squared Error:",mean_squared_error(y_test,y_pred))
print("Root Mean Square Error:",math.sqrt(mean_squared_error(y_test,y_pred)))
print("Mean Absolute Error:",mean_absolute_error(y_test,y_pred))
print("R2 Score :",r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
plt.scatter(y_test,y_pred)
```

```python
sns.distplot((y_test-y_pred),bins=50)
```

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 150, random_state =0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Initialize the Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=0)

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Make predictions
y_pred_rf = best_model.predict(X_test)

# Evaluate the model
r2_rf = r2_score(y_test, y_pred_rf)
print("R-squared Score (Random Forest):", r2_rf)
```
